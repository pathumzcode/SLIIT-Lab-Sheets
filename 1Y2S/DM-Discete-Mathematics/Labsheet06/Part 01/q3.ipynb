. The harmonic sum is the sum of reciprocals of the positive integers.



Write a recursive Python program to calculate the sum of harmonic series
harmonic_sum(n) up to n terms. Use your function to compute and print:
harmonic_sum(5) = 2.2833.

```python
def harmonic_sum(n):
    if n == 1:
        return 1
    else:
        return 1/n + harmonic_sum(n-1)

print(harmonic_sum(5))
```
