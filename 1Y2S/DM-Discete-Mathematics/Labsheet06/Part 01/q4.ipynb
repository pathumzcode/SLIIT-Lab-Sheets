A biologist is studying a special species of bacteria. The bacteria population grows in a
unique pattern.
On day 1, the population is 2, and on day 2, it is 3. From day 3 onwards, the population on
any day is equal to the product of the populations on the previous two days.
Write a recursive formula and develop a Python function bacteria(n) that returns the
population on day n.

```python
def  bacteria(n):
    if n > 0 and  n <= 2:
        return n+1
    else:
        return bacteria(n-1) + bacteria (n-2)


print(bacteria(1))
print(bacteria(5))
```
