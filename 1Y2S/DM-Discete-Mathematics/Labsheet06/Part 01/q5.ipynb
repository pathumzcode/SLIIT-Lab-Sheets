In Pascal’s Triangle.
• Computation starts with the value 1 at the top level.
• Each value in the structure is computed as the sum of the two values directly above
it.
• The value at row n and column k is represented by 𝒑𝒂𝒔𝒄𝒂𝒍(𝒏, 𝒌).

Base case: 𝒑𝒂𝒔𝒄𝒂𝒍(𝒏, 𝟎) = 𝟏 & 𝒑𝒂𝒔𝒄𝒂𝒍(𝒏,𝒏) = 𝟏 for all non-negative integers n.
Recursive Step:
𝒑𝒂𝒔𝒄𝒂𝒍(𝒏, 𝒌) = 𝒑𝒂𝒔𝒄𝒂𝒍(𝒏 − 𝟏, 𝒌 − 𝟏) + 𝒑𝒂𝒔𝒄𝒂𝒍(𝒏 − 𝟏, 𝒌)
Write a recursive function pascal(n, k) that returns the value at row n and column k.

```python
def pascal(n, k):
    if (k == 0 or k == n):
        return 1
    else:
        return pascal(n-1, k-1) + pascal(n-1, k)

print(pascal(1, 0))
print(pascal(1, 1))
print(pascal(7, 6))
```
