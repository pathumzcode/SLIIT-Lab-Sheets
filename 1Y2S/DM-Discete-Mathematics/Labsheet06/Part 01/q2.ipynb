In a logging system, the total number of log entries processed after n cycles is defined by
the recursive relation:
𝑳(𝒏) = 𝑳(𝒏 − 𝟏) + 𝟓
Base case: 𝐿(0) = 10
Write a Python function L(n) that uses recursion to compute the number of log entries after
n cycles. Use your function to compute and print the values of: 𝐿(1), 𝐿(2), 𝑎𝑛𝑑 𝐿(3)


```python
def L(n):
    if n==0:
        return 10
    else:
        return L(n-1) + 5

print(L(1))
print(L(2))
print(L(3))
```
