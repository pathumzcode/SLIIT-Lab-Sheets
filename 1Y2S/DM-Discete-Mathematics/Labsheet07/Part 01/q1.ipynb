A small airline company maps its flight routes as a directed graph. Cities are vertices, and
flights are directed edges. The company wants to analyze this graph for optimization. Cities
(Vertices): A, B, C, D, E Flights (Edges): (A → B), (B → C), (C → D), (C → E), (E → D), (C → B)
I. Write a Python program to Represent this graph as both an adjacency list and an
adjacency matrix.
II. Display the graph using a matplotlib library.

```python
import networkx as nx
import matplotlib.pyplot as plt

airline = nx.DiGraph()
cities = (['A', 'B', 'C', 'D', 'E'])
flights = [('A', 'B'), ('A', 'C'), ('B', 'C'), ('C', 'D'), ('D', 'E')]

airline.add_nodes_from(cities)
airline.add_edges_from(flights)

print("Cities:", airline.nodes())
print("Flights:", airline.edges())
print(airline)

nx.draw(airline, with_labels=True)
plt.show()

adj_list = nx.to_dict_of_lists(airline)
print(adj_list)
```
