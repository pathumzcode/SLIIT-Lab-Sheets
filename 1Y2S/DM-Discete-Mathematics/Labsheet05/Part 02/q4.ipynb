A cloud-based distributed computing system consists of 15 distinct servers, classified as
follows.
• 8 computer servers (used for processing tasks)
• 5 storage servers (used for data storage)
• 2 backup servers (used for fault tolerance and recovery)
In a large-scale deployment environment, a system administrator is required to activate 11
servers. Write a Python program to accomplish the following tasks.
a) If there are no restrictions on the types of servers selected, how many different sets of
11servers can be activated?
b) If the deployment policy requires exactly 6 compute servers, 4 storage servers, and 1
backup server, how many different valid selections are possible?

```python
import math

total_servers = 15
activate_count = 11

# a) no restrictions
total_sets = math.comb(total_servers, activate_count)

# b) exact composition
compute_servers = 8
storage_servers = 5
backup_servers = 2

valid_sets = (
    math.comb(compute_servers, 6)
    * math.comb(storage_servers, 4)
    * math.comb(backup_servers, 1)
)

print("a) Total sets without restrictions:", total_sets)
print("b) Total valid selections with 6 compute, 4 storage, 1 backup:", valid_sets)
```
