A cloud service provider offers 4 types of virtual machines (VMs), Compute-optimized,
Memory-optimized, Storage-optimized, and GPU-based. A system administrator needs to
allocate 10 virtual machines to a project. Assume that,
• Multiple VMs of the same type can be allocated.
• Only the number of VMs of each type matters.
• The order in which VMs are allocated is irrelevant.
Write a Python program to determine how many distinct VM allocation combinations are
possible.


```python
import math

types = 4
vms = 10
combinations = math.comb(vms + types - 1, types - 1)
print(combinations)
```
