 A theater company wants to schedule performances with a specific set of actors. Each
performance involves arranging the actors in a unique order.
a) Generate all possible schedules for 3 actors ("A", "B", "C").
b) Find the 5th schedule in the list of all possible schedules.


```python
#Part A
import itertools

actors = ['A','B','C']

nPr_1 = itertools.permutations(actors, 3)

print(list(nPr_1))
```

```python
from itertools import permutations

actors = ['A','B','C']

nPr_1 = list(permutations(actors, 3))

print(nPr_1)
```

```python
# Part B
print(nPr_1[4])
```
