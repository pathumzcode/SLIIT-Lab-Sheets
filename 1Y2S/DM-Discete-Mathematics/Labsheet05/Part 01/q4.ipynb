A team leader wants to form committees from a group of employees for a project. Each
committee must have exactly 2 members.
a) List all possible committees from employees ["A", "B", "C", "D"].
b) Find the 3rd committee on the list of all possible committees

```python
#Part A
from itertools import combinations

employees = ['A', 'B', 'C', 'D']

nCr = list(combinations(employees, 2))

print(nCr)

#Part A
print(nCr[2])
```
