A computer science student designs passwords. The passwords must consist of:
• 2 lowercase letters (chosen from ["a", "b", "c", "d"]).
• 3 digits (chosen from 5 digits: 0 to 4).


```python
from itertools import product

letters = ['a', 'b', 'c', 'd']
digits = ['0', '1', '2', '3', '4']

all_chars = letters + digits

nPr_2 = list(product(all_chars, repeat=5))

valid_passwords = []

i = 0

while i < len(nPr_2):

    letters_count = 0
    digits_count = 0

    for j in nPr_2[i]:
        if j in letters:
            letters_count += 1
        elif j in digits:
            digits_count += 1
    if letters_count == 2 and digits_count == 3:
        valid_passwords.append(nPr_2[i])
    i += 1
    
print(valid_passwords)

print(len(valid_passwords))
```
