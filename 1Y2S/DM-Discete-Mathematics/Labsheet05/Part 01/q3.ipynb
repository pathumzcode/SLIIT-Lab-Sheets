A robot needs to navigate a cartesian plane (XY plane) starting at the origin (0,0) and reaching
the up-right corner (3,3). The robot can only move right (R) or up (U).
a) Write a Python program to calculate the number of unique paths.
b) Display one valid path.

```python
import math

rows, cols = 3, 3
total_moves = rows + cols

unique_paths = math.comb(total_moves, rows)  # or math.comb(total_moves, cols)
one_path = "R" * cols + "U" * rows

print("Number of unique paths:", unique_paths)
print("One valid path:", one_path)
```
