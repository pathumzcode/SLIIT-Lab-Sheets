A florist is designing bouquets using flowers of 3 types. Each bouquet must have 3 flowers,
and repetition of flower types is allowed.
a) List all possible bouquets using ["Rose", "Lily", "Tulip"].
b) Display the bouquets where all three flowers are different

```python
from itertools import product

flowers = ["Rose", "Lily", "Tulip"]

bouquets = [list(p) for p in product(flowers, repeat=3)]
all_diff = [b for b in bouquets if len(set(b)) == 3]

print("All possible bouquets:")
for b in bouquets:
    print(b)

print("\nBouquets with all different flowers:")
for b in all_diff:
    print(b)
```
