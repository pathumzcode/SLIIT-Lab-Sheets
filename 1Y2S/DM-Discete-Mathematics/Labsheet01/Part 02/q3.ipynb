Create Student Grade Management System:  
• Ask for student’s name. 
• Take marks for 3 subjects. 
• Calculate the total mark out of 100 (percentage) and print detailed mark sheet with 
all information.  
Determine grade as below. 
   - A: >=75% 
   - B: 60-75% 
   - C: 45-60% 
   - D: 30-45% 
   - F: below 30% 

```python
name = str(input("Enter your name: "))

total = 0
average = 0

for i in range (1,4):
    marks = float(input("Enter your marks for subject " + str(i) + ": "))
    total += marks

average = total/3

print("Student Name: ", name)
print("Total: ", total)
print("Average: ", average)

if (average >= 75):
    print("Grade: A")
elif (average >= 60):
    print("Grade: B")
elif (average >= 40):
    print("Grade: C")
else:
    print("Grade: F")
```
