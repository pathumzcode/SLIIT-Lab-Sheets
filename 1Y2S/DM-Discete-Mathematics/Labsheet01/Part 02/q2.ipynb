 Create a program to simulate BMI calculator: 
• Asks user for weight (kg) and height (m) 
• Calculate BMI using formula: weight / (height × height) 
• Classifies BMI (Underweight: <18, Normal: 18-25, Overweight: >25) 
• Print BMI value and classification

```python
height = float(input("Enter your height (m) : "))
weight = float(input("Enter your weight (kg): "))

BMI =  weight/(height*height)

print("Your BMI is : ", BMI)

if BMI < 18.0:
    print("Underweight")
elif BMI >= 18.0 and BMI < 25.0:
    print("Normal")
elif BMI >= 25.0:
    print("Overweight")
```
