3. Create a program to simulate simple calculator: 
• Asks user for two numbers 
• Asks user for operation (+, -, ×, ÷) 
• Print the result with appropriate message. 

```python
num1 = int(input("Enter 1st number : "))
num2 = int(input("Enter 2nd number : "))

operation = input("Enter the operator (+, -, *, /): ")

if operation == "+":
    print(num1 + num2)
elif operation == "-":
    print(num1 - num2)
elif operation == "*":
    print(num1 * num2)
elif operation == "/":  
    if num2 != 0:       
        print(num1 / num2)
    else:
        print("Cannot divide by zero")
else:
    print("Invalid operator entered!")
```
