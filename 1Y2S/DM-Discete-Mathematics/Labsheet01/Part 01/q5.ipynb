5. Write a Python program that continuously asks the user to enter a password using a while 
loop. The program should keep requesting the password until the correct password 
"python123" is entered. Once the correct password is entered, display the message 
"Access granted". 

```python
while True:
    password = str(input("Enter the password: "))
    if password == "python123":
        print("Access granted")
        break
    else:
        print("Access denied")
```
