2. Write Python Program to check Validity of Triangle. Ask the user for three sides of a 
triangle, check if they form a valid triangle, and print the result. (Hint: - Triangle Inequality 
Theorem: The sum of the lengths of any two sides must be greater than the length of the 
third side.) 

```python
side01 = int(input("Input side 01: "))
side02 = int(input("Input side 02: "))
side03 = int(input("Input side 03: "))

if (side01 + side02 > side03) and (side01 + side03 > side02) and (side02 + side03 > side01):
    print("The sides can form a triangle.")
else:
    print("The sides cannot form a triangle.")
```
